## YallaKora

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL of the match center
url = 'https://www.yallakora.com/Match-Center/?date=18/21/2023'
```

```python
# Send an HTTP GET request to the URL
response = requests.get(url)
```

```python
# Get response content
html_content = response.content
print(html_content)
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Find all match elements on the page
championships = soup.find_all('div', class_='matchCard')
```

```python
# Get the first championship title
championship_title = championships[0].contents[1].find('h2').text.strip()
championship_title
```

```python
# Get all the championship titles
for i in range(len(championships)):
    print(championships[i].contents[1].find('h2').text.strip())
```

```python
# Get the first championship matches
matches = championships[0].contents[3].find_all('div', class_='teamCntnr')
matches
```

```python
# Get the first championship matches detailes 
match_details = []

for i in range(len(matches)):
    teamA = matches[i].find('div', {'class':'teamA'}).text.strip()
    teamB = matches[i].find('div', {'class':'teamB'}).text.strip()

    results = matches[i].find('div', {'class':'MResult'}).find_all('span', {'class':'score'})
    score = f'{results[0].text.strip()} - {results[1].text.strip()}'

    time = matches[i].find('div', {'class':'MResult'}).find('span', {'class':'time'}).text.strip()

    print(teamA, teamB, score, time)
```

```python
# Get all information about all the championships
for i in range(len(championships)):
    print(championships[i].contents[1].find('h2').text.strip())

    # printing the number of games played and goals scored by each team in every match
    matches = championships[i].contents[3].find_all('div', class_='teamCntnr')
    for j in range(len(matches)):
        teamA = matches[j].find('div', {'class':'teamA'}).text.strip()
        teamB = matches[j].find('div', {'class':'teamB'}).text.strip()

        results = matches[j].find('div', {'class':'MResult'}).find_all('span', {'class':'score'})
        score = f'{results[0].text.strip()} - {results[1].text.strip()}'

        time = matches[j].find('div', {'class':'MResult'}).find('span', {'class':'time'}).text.strip()

        print(f"{teamA} {score} {teamB}\n\t{time}")
    
    print('='*20)
```

## Companies Revenue

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

headers = {
    "User-Agent": "Mozilla/5.0 (compatible; MyBot/1.0; +https://example.com/bot)"  
}

resp = requests.get(url, headers=headers)
print(resp.status_code)   # لازم تكون 200

```

```python

soup = BeautifulSoup(resp.content, 'html.parser')
```

```python
print(resp.text[:500])
```

```python
# Find all match elements on the page
all_tables = soup.find_all("table")
print("عدد الجداول:", len(all_tables))
```

```python

# find table (wanted)
for idx, tab in enumerate(all_tables[:4]):
    # print tables headers 
    headers = [th.get_text(strip=True) for th in tab.find_all('th')]
    print(idx, headers)
```

```python
# لنفترض أنك رأيت أن الجدول الذي تريده هو all_tables[1]:
target = all_tables[1]

companies = []
for row in target.find_all('tr')[1:]:  # تخطّي أول صف العناوين
    cols = row.find_all(['td','th'])
    if len(cols) >= 6:
        company = cols[1].get_text(strip=True)
        revenue = cols[3].get_text(strip=True)
        employees = cols[5].get_text(strip=True)
        companies.append({'Company': company, 'Revenue': revenue, 'Employees': employees})

print(companies[:5])
```

```python
# Print the collected data
for company in companies:
    print(f"Company: {company['Company']}")
    print(f"Revenue: {company['Revenue']}")
    print(f"Employees: {company['Employees']}")
    print("----------")
```

## Book Shop

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define the base URL and starting page number
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
page_number = 1
# Loop through multiple pages
while True:
    url = base_url.format(page_number)

    # Send an HTTP GET request to the URL
    response = requests.get(url)
    if response.status_code != 200:
        break

    html_content = response.content

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all book article elements on the page
    books = soup.find_all('article', class_='product_pod')

    # Loop through each book and extract relevant information
    for book in books:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        
        rating = 'No rating' if 'star-rating' not in str(book) else book.find('p', class_='star-rating')['class'][-1]

        print(f"Title: {title}")
        print(f"Price: {price}")
        print(f"Rating: {rating}")
        print("----------")

    # # Move to the next page
    page_number += 1
```
